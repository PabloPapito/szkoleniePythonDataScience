# Wizualizacja Danych z Matplotlib i Seaborn

## 1. Rodzaje Wykresów i Wizualizacji:

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Przykładowe dane
data = {'Category': ['A', 'B', 'C', 'D'],
        'Value': [30, 50, 20, 40]}

df = pd.DataFrame(data)

# Wykres słupkowy w Matplotlib
plt.bar(df['Category'], df['Value'])
plt.title('Wykres słupkowy - Matplotlib')
plt.xlabel('Kategoria')
plt.ylabel('Wartość')
plt.show()

# Wykres słupkowy w Seaborn
sns.barplot(x='Category', y='Value', data=df)
plt.title('Wykres słupkowy - Seaborn')
plt.xlabel('Kategoria')
plt.ylabel('Wartość')
plt.show()
```

## 2. Elementy Storytelling w Wizualizacji:

```python
# Wykres z elementami storytellingu - Seaborn
sns.barplot(x='Category', y='Value', data=df, palette='viridis')
plt.title('Wartość w kategoriach')
plt.xlabel('Kategoria')
plt.ylabel('Wartość')
plt.show()
```

## Rozwiązanie

```python
# Rozwiązanie: Tworzenie interaktywnych wykresów
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interactive

# Przykładowe dane
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

def interactive_plot(a=1.0, b=1.0):
    # Funkcja tworząca interaktywny wykres
    plt.figure(figsize=(8, 5))
    plt.plot(x, a * np.sin(b * x), label=f'{a} * sin({b} * x)')
    plt.plot(x, a * np.cos(b * x), label=f'{a} * cos({b} * x)')
    plt.title(f'Interaktywny Wykres Sinusa i Cosinusa')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.legend()
    plt.grid(True)
    plt.show()

# Utworzenie interaktywnego wykresu
interactive_plot_widget = interactive(interactive_plot, a=(0.1, 2.0, 0.1), b=(0.1, 2.0, 0.1))

# Wyświetlenie interaktywnego widgetu
interactive_plot_widget
```

## 4. Przydatne Funkcje i Skróty:

```python
# Przykład użycia przydatnych funkcji
plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='cos(x)')
plt.title('Wykres Sinusa i Cosinusa')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()
```

## Rozwiązanie

```python
# Rozwiązanie: Eksperymentowanie z różnymi wykresami
import numpy as np

# Przykładowe dane
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Wykres liniowy
plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
plt.plot(x, y, label='sin(x)')
plt.title('Wykres Liniowy')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

# Wykres punktowy
plt.subplot(2, 2, 2)
plt.scatter(x, y, label='sin(x)', color='red')
plt.title('Wykres Punktowy')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

# Histogram
plt.subplot(2, 2, 3)
plt.hist(y, bins=20, color='green', alpha=0.7)
plt.title('Histogram')
plt.xlabel('Wartość')
plt.ylabel('Liczebność')

# Wykres słupkowy
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 1, 5]
plt.subplot(2, 2, 4)
plt.bar(categories, values, color='purple')
plt.title('Wykres Słupkowy')
plt.xlabel('Kategoria')
plt.ylabel('Wartość')

plt.tight_layout()
plt.show()
```
