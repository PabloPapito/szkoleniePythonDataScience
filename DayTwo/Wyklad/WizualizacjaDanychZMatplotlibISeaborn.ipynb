# Wizualizacja Danych z Matplotlib i Seaborn

## 1. Rodzaje Wykres√≥w i Wizualizacji:

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Przyk≈Çadowe dane
data = {'Category': ['A', 'B', 'C', 'D'],
        'Value': [30, 50, 20, 40]}

df = pd.DataFrame(data)

# Wykres s≈Çupkowy w Matplotlib
plt.bar(df['Category'], df['Value'])
plt.title('Wykres s≈Çupkowy - Matplotlib')
plt.xlabel('Kategoria')
plt.ylabel('Warto≈õƒá')
plt.show()

# Wykres s≈Çupkowy w Seaborn
sns.barplot(x='Category', y='Value', data=df)
plt.title('Wykres s≈Çupkowy - Seaborn')
plt.xlabel('Kategoria')
plt.ylabel('Warto≈õƒá')
plt.show()
```

## 2. Elementy Storytelling w Wizualizacji:

```python
# Wykres z elementami storytellingu - Seaborn
sns.barplot(x='Category', y='Value', data=df, palette='viridis')
plt.title('Warto≈õƒá w kategoriach')
plt.xlabel('Kategoria')
plt.ylabel('Warto≈õƒá')
plt.show()
```

## RozwiƒÖzanie

```python
# RozwiƒÖzanie: Tworzenie interaktywnych wykres√≥w
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interactive

# Przyk≈Çadowe dane
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

def interactive_plot(a=1.0, b=1.0):
    # Funkcja tworzƒÖca interaktywny wykres
    plt.figure(figsize=(8, 5))
    plt.plot(x, a * np.sin(b * x), label=f'{a} * sin({b} * x)')
    plt.plot(x, a * np.cos(b * x), label=f'{a} * cos({b} * x)')
    plt.title(f'Interaktywny Wykres Sinusa i Cosinusa')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.legend()
    plt.grid(True)
    plt.show()

# Utworzenie interaktywnego wykresu
interactive_plot_widget = interactive(interactive_plot, a=(0.1, 2.0, 0.1), b=(0.1, 2.0, 0.1))

# Wy≈õwietlenie interaktywnego widgetu
interactive_plot_widget
```

## 4. Przydatne Funkcje i Skr√≥ty:

```python
# Przyk≈Çad u≈ºycia przydatnych funkcji
plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='cos(x)')
plt.title('Wykres Sinusa i Cosinusa')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()
```

## RozwiƒÖzanie

```python
# RozwiƒÖzanie: Eksperymentowanie z r√≥≈ºnymi wykresami
import numpy as np

# Przyk≈Çadowe dane
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Wykres liniowy
plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
plt.plot(x, y, label='sin(x)')
plt.title('Wykres Liniowy')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

# Wykres punktowy
plt.subplot(2, 2, 2)
plt.scatter(x, y, label='sin(x)', color='red')
plt.title('Wykres Punktowy')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

# Histogram
plt.subplot(2, 2, 3)
plt.hist(y, bins=20, color='green', alpha=0.7)
plt.title('Histogram')
plt.xlabel('Warto≈õƒá')
plt.ylabel('Liczebno≈õƒá')

# Wykres s≈Çupkowy
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 1, 5]
plt.subplot(2, 2, 4)
plt.bar(categories, values, color='purple')
plt.title('Wykres S≈Çupkowy')
plt.xlabel('Kategoria')
plt.ylabel('Warto≈õƒá')

plt.tight_layout()
plt.show()
```
